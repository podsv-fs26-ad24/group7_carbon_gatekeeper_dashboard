# 1 Evaluation Report

This report documents the systematic evaluation of the visualization prototype developed in the [Visual Encoding and Design](https://podsv-fs26-ad24.github.io/ad24-7-fancyproject/viz_design_report.html) phase against the project objectives and success criteria defined in the [Project Charta](https://podsv-fs26-ad24.github.io/ad24-7-fancyproject/project_charta.html).

## 1.1 Evaluation Setup

## 1.1.1 Evaluation Objectives

The evaluation aimed to assess whether the developed Business Travel CO2 Dashboard successfully supports the analytical, communicative and operational goals defined in the Project Charta. The prototype was evaluated against the identified success criteria from the visual encoding and design phase, focusing on its ability to support sustainability-oriented travel decision making for organizational stakeholders.

The evaluation specifically examined whether the dashboard enables users to monitor CO2 budget compliance, identify reduction opportunities, compare business units and interpret travel-related emission patterns efficiently and accurately.

The assessment was structured according to three value dimensions commonly used in visualization evaluation.

### Cognitive and Analytical Value

The cognitive and analytical evaluation focused on whether users were able to effectively accomplish the intended analytical tasks using the dashboard. These tasks included:

- detecting CO2 budget overruns and approaching thresholds,
- comparing emission performance across business units,
- identifying high-emission travel routes,
- detecting reduction opportunities through alternative transport modes,
- interpreting geographic travel distributions,
- evaluating scenario changes between planned and optimized travel configurations.

The evaluation further examined whether the selected visual encodings supported rapid pattern recognition, comparison tasks and insight generation without causing unnecessary cognitive load.

### Communicative Value

The communicative evaluation examined whether the visualization successfully conveyed sustainability-related insights and recommendations to the target audience, including users with varying levels of data literacy.

Particular attention was given to:

- clarity of KPI communication,
- interpretability of gauges, maps and comparison charts,
- effectiveness of semantic colour encoding,
- understandability of reduction recommendations,
- consistency of labels, terminology and visual hierarchy,
- narrative flow across dashboard sections.

The evaluation assessed whether users could understand the dashboard’s main messages without requiring extensive prior knowledge of data visualization techniques.

### Experiential and Aesthetic Value

The experiential and aesthetic evaluation focused on perceived usability, trustworthiness and engagement of the dashboard interface.

This included evaluating:

- overall usability and navigation,
- consistency of the visual design system,
- perceived professionalism and credibility,
- readability and accessibility,
- effectiveness of layout and whitespace usage,
- interaction quality of filtering, drill-down and scenario switching features,
- overall user confidence while exploring sustainability insights.

The evaluation also examined whether the dashboard supported a smooth analytical workflow while maintaining a visually coherent and uncluttered interface.

---

## 1.1.2 Evaluation Method

The evaluation combined heuristic evaluation, insight-based evaluation and comparative evaluation methods in order to assess both the analytical effectiveness and usability of the visualization prototype.

### Heuristic Evaluation

A heuristic evaluation was conducted based on established visualization and usability principles from Munzner, Few, Tufte and Nielsen. The dashboard design was systematically reviewed with regard to:

- visual hierarchy,
- consistency of visual encodings,
- semantic colour usage,
- reduction of visual clutter,
- cognitive load management,
- dashboard layout organization,
- interaction design,
- readability and accessibility.

Special attention was given to whether the dashboard followed recognized dashboard design guidelines such as KPI prioritization, progressive disclosure of detail and consistency across visual components.

### Insight-Based Evaluation

An insight-based evaluation was performed to assess whether the dashboard enabled meaningful analytical insights within a short interaction time.

The evaluation examined whether users could:

- identify emission hotspots,
- detect business units approaching CO2 limits,
- recognize reduction opportunities,
- compare travel scenarios,
- derive actionable sustainability recommendations.

The quality and interpretability of generated insights were assessed through structured walkthroughs of realistic travel planning scenarios.

### Comparative Evaluation

A comparative evaluation was conducted by comparing the original “as planned” travel scenario with the optimized scenario generated through suggested transport mode alternatives.

This evaluation examined:

- visibility of CO2 reduction potential,
- interpretability of optimization effects,
- usefulness of scenario switching,
- support for sustainability-oriented decision making.

The comparison allowed the evaluation of how effectively the dashboard communicates the impact of alternative travel decisions.

### Task-Based Walkthrough

In addition to heuristic inspection, representative analytical tasks were performed using the dashboard prototype. These tasks included:

1. Identifying business units approaching their CO2 budget limit.
2. Detecting routes with the highest reduction potential.
3. Comparing transport alternatives for selected routes.
4. Interpreting global travel emission distributions.
5. Evaluating the effect of applying alternative transportation modes.

The walkthrough evaluated whether the dashboard structure and interaction flow supported efficient task completion.

The primary evaluators were the project team members responsible for the design and implementation of the dashboard prototype. The evaluation incorporated perspectives related to data visualization design, sustainability analytics and dashboard usability.

The evaluation focused on representative target users identified during the user analysis phase, including:

- sustainability managers,

## 1.2 Results

## 1.2.1 Cognitive and Analytical Results

The dashboard successfully supported the intended analytical tasks and enabled efficient exploration of travel-related CO2 data.

### Budget Monitoring and Threshold Detection

Users were able to rapidly identify overall budget utilisation and detect business units approaching their annual CO2 limits through the KPI overview section and the business unit gauges.

The combined use of:

- KPI summary cards,
- semantic threshold colouring,
- gauge indicators,
- and budget comparison bars

enabled immediate recognition of critical emission levels.

The dashboard showed that:

- total emissions reached 802.3 t CO2,
- corresponding to 95% budget utilisation,
- with a remaining margin of 38.8 t below the total allocated budget.

Business units BU2, BU3 and BU4 were identified as approaching their respective budget limits, with utilisation levels between 95% and 98%.

The evaluation demonstrated that the dashboard effectively supported rapid status assessment and comparative analysis across organizational units.

### Identification of Reduction Opportunities

The reduction lever section enabled users to identify routes with significant CO2 reduction potential through alternative transportation modes.

The dashboard detected:

- 404.6 t total CO2 reduction potential,
- across 22 major travel routes,
- primarily through replacing short-haul flights with train connections.

The ranked recommendation table supported prioritisation by displaying:

- number of affected flights,
- average emissions per route,
- estimated savings,
- percentage reduction potential.

The strongest reduction opportunities were identified on European short-haul routes such as:

- Amsterdam ↔ Zurich,
- Vienna ↔ Zurich,
- Brussels ↔ Zurich,
- Berlin ↔ Zurich.

The scenario comparison mechanism further enabled users to evaluate the impact of applying transportation alternatives directly within the dashboard.

### Geographic Pattern Recognition

The geographic visualization successfully communicated global travel concentration patterns and emission-intensive flight corridors.

Users were able to identify:

- strong concentration of routes around Central Europe,
- intercontinental travel clusters,
- dominant dependence on air travel,
- spatial distribution of organizational travel behaviour.

The map interaction supported exploratory analysis through hover-based route inspection and regional filtering.

### Route-Level Analysis

The route deep-dive section enabled detailed comparison of transportation alternatives for selected routes.

Users could compare:

- CO2 emissions,
- estimated travel duration,
- reduction percentages,
- and transport modes

within a single analytical view.

This supported scenario-based decision making and improved interpretability of trade-offs between sustainability and travel time.

## 1.3.2 Areas Where the Design Fell Short

Despite the positive overall evaluation, several limitations and improvement opportunities were identified.

### Complexity of Advanced Analytical Views

Although the dashboard structure supported efficient navigation, some advanced analytical sections required moderate analytical literacy.

In particular:

- route deep-dive comparisons,
- transport mode trade-off analysis,
- and interpretation of scenario optimization results

may be difficult for inexperienced users without additional onboarding or explanation.

The prototype prioritised analytical depth over complete beginner accessibility in some sections.

### Dependence on Historical Route Data

The recommendation system relied heavily on historical travel averages and previously observed transportation modes.

This introduces several limitations:

- recommendations may not always reflect current transportation availability,
- estimated travel durations may differ from real operational conditions,
- route recommendations are constrained by the historical dataset,
- exceptional contextual factors are not represented.

As a result, the optimization logic should be interpreted as analytical decision support rather than fully automated operational planning.

### Limited Accessibility Features

The current prototype did not yet include advanced accessibility support such as:

- screen reader optimisation,
- keyboard-only navigation,
- colourblind-specific modes,
- responsive mobile adaptation.

While the visual design remained relatively clean and readable, accessibility considerations would require further refinement before large-scale deployment.

### Potential Information Density

Although progressive disclosure reduced much of the complexity, the dashboard still contains a large amount of analytical information within a single page.

Users primarily interested in high-level reporting may not require all available analytical detail.

Future iterations could therefore consider:

- role-specific dashboard views,
- simplified executive summaries,
- or customizable analytical modules.

---

## 1.3.3 Trade-Offs Between the Value Dimensions

The evaluation revealed several trade-offs between analytical richness, accessibility and interaction complexity.

### Analytical Depth vs Accessibility

The dashboard intentionally prioritised analytical capability and actionable sustainability insights.

Features such as:

- route-level comparison,
- scenario simulation,
- and reduction potential analysis

provided strong analytical value but increased interpretive complexity for less experienced users.

The design attempted to mitigate this trade-off through:

- semantic colour encoding,
- KPI-first structure,
- progressive disclosure,
- and explanatory section subtitles.

Nevertheless, complete accessibility for users with very low analytical literacy may require additional simplification.

### Information Density vs Dashboard Compactness

The single-page dashboard structure enabled efficient navigation and reduced context switching.

However, integrating multiple analytical perspectives within one interface also increased information density.

The project therefore balanced:

- overview visibility,
- detail accessibility,
- and interaction efficiency

against risks of cognitive overload.

The current implementation achieved a generally effective compromise, although additional modularisation could further improve usability.

### Precision vs Practical Decision Support

The optimization system prioritised practical decision support over exact real-world simulation accuracy.

For example:

- estimated travel durations,
- CO2 averages,
- and route recommendations

were intentionally simplified in order to support rapid interpretation and usability.

This trade-off improved communicative clarity while limiting predictive precision.

While sufficient for analytical prototyping, these assumptions limit the precision of operational recommendations.
